```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.figure(figsize=(12,6))

plt.scatter(x='Year', y='CSIRO Adjusted Sea Level', data =df, cmap='red')
x=pd.Series(range(1880,2050))
plt.plot(x, slope*x+intercept)
```

```python
df2 = df.loc[df['Year']>=2000]
slope1, intercept1, r_value1, p_value1, std_err1 = stats.linregress(df2['Year'], df2['CSIRO Adjusted Sea Level'])
```

```python
plt.figure(figsize=(12,6))
fig, ax = plt.subplots()
x=pd.Series(range(2000,2050))
ax.plot(x, slope1*x+intercept1)
ax.set(xlabel="Year", ylabel="Sea Level (inches)", title="Rise in Sea Level")
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
plt.figure(figsize=(12,6))

slope, intercept, r_value, p_value, std_err = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])
years = pd.Series(range(1880,2051))

plt.plot(years, intercept + slope*years, 'r', label='first line of best fit')
plt.scatter(x='Year', y='CSIRO Adjusted Sea Level', data =df, cmap='red')
```
