```python
df_heat = df[ 
        ( df['ap_lo'] <= df['ap_hi'] ) & 
        ( df['height'] >= df['height'].quantile(0.025) ) & 
        ( df['height'] <= df['height'].quantile(0.975) ) & 
        ( df['weight'] >= df['weight'].quantile(0.025) ) & 
        ( df['weight'] <= df['weight'].quantile(0.975) ) 
    ]
```

```python
df_heat.head()
```

```python
corrDf= df_heat.corr()
```

```python
corrDf
```

```python
final_map_masked = np.triu(corrDf)
```

```python
plt.figure(figsize=(5,6))
```

```python
fig, ax = plt.subplots(figsize=(10,10))
```

```python
fig, ax = plt.subplots(figsize=(12,12))
ax = sns.heatmap(corrDf, mask=final_map_masked, annot=True, fmt='0.1f', square=True, cmap = 'rocket')
```
