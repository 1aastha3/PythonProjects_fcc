```python
import numpy as np

def calculate(list):

  if(len(list)!=9):
    raise ValueError('List must contain nine numbers.')
    
  arr = np.array(list)
  arr1 = np.reshape(arr, (3,3))
  output = {}

  output['mean'] = [arr1.mean(axis =0).tolist(),       arr1.mean(axis=1).tolist(), arr1.mean().tolist()]

  output['variance'] = [arr1.var(axis=0).tolist(), arr1.var(axis=1).tolist(), arr1.var().tolist()]
  
  output['standard deviation'] = [arr1.std(axis=0).tolist(), arr1.std(axis=1).tolist(), arr1.std().tolist()]

  output['max']= [arr1.max(axis=0).tolist(), arr1.max(axis=1).tolist(), arr1.max().tolist()]

  output['min'] = [arr1.min(axis=0).tolist(), arr1.min(axis=1).tolist(), arr1.min().tolist()]

  output['sum'] = [arr1.sum(axis=0).tolist(), arr1.sum(axis=1).tolist(),arr1.sum().tolist()]
    


  return output

```

```python
calculate([0,1,2,3,4,5,6,7,8])
```

```python
calculate([0,9,8])
```
